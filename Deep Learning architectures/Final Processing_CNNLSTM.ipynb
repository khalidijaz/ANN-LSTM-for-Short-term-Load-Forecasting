```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
Processed_data =pd.read_csv("Book_1.csv")
Processed_data
```

```python
Processed_data=pd.DataFrame(Processed_data)
Processed_data.info()
```

```python
Processed_data['Temperature']=Processed_data['Temperature'].astype(np.float64)
Processed_data['Holidays    0 for close   1 for open']=Processed_data['Holidays    0 for close   1 for open'].astype(np.float64)
Processed_data['year']=Processed_data['year'].astype(np.float64)
Processed_data['month']=Processed_data['month'].astype(np.float64)
Processed_data['day']=Processed_data['day'].astype(np.float64)
Processed_data['hour']=Processed_data['hour'].astype(np.float64)
Processed_data['minute']=Processed_data['minute'].astype(np.float64)
Processed_data['second']=Processed_data['second'].astype(np.float64)
Processed_data['weekday']=Processed_data['weekday'].astype(np.float64)
Processed_data['Load']=Processed_data['Load'].astype(np.float64)
Processed_data.info()
```

```python
train= Processed_data.copy()[['Temperature','Humidity', 'Holidays    0 for close   1 for open','year', 'month', 'day', 'hour', 'minute', 'second', 'weekday','Load']]
```

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()
train[['Temperature','Humidity', 'Holidays    0 for close   1 for open','year', 'month', 'day', 
       'hour', 'minute', 'second', 'weekday','Load']] = Scaler.fit_transform(Processed_data)
train
```

```python
train_shifted = train.copy()
train_shifted = train_shifted.astype(np.float64)
train_shifted['y_t+1'] = train_shifted['Load'].shift(-1)
train_shifted.head(10)
```

```python
train_shifted = train_shifted.rename(columns={'Load':'Original_Load'})
train_shifted.head(10)
```

```python
train_shifted = train_shifted.dropna(how='any')
train_shifted.head(5)
```

```python
X= train_shifted[['Temperature','Humidity', 'Holidays    0 for close   1 for open','year', 'month', 'day', 'hour', 'minute', 'second', 'weekday','Original_Load']]
X
```

```python
Y= train_shifted[['y_t+1']]
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.6,test_size=0.2,random_state = 10, shuffle=False)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import models
from keras.applications import DenseNet121
from keras.layers.core import Dense, Activation, Flatten
from keras.layers.recurrent import LSTM,GRU
```

```python
X_train=np.atleast_3d(X_train)
X_train.shape
```

```python
from keras.layers import BatchNormalization
from keras.layers import Input
from keras.layers.convolutional import Conv1D, MaxPooling1D,MaxPooling2D
model = models.Sequential()
model.add(Conv1D(filters=128, kernel_size=1, activation='relu',input_shape=(11,1)))
#model.add(TimeDistributed(MaxPooling1D(pool_size=1,strides=1, padding='valid')))
#model.add(TimeDistributed(MaxPooling1D(pool_size=1,strides=1, padding='valid')))
#model.add(Dropout(0.1))
model.add(MaxPooling1D(pool_size=1,strides=1, padding='valid'))
#model.add(Dropout(0.1))
#model.add(TimeDistributed(MaxPooling1D(pool_size=2,strides=1, padding='valid')))
#model.add(Dropout(0.1))
#model.add(Flatten())
#model.add(RepeatVector(5629))
model.add((LSTM(8,activation='linear')))
#model.add(Flatten())
#model.add(GRU(300))
#model.add(Dense(150,activation='linear'))
#model.add(Dense(75,activation='linear'))
#model.add(Dropout(0.1))
#model.add(Bidirectional(LSTM(300)))
model.add(Dense(1,activation='linear'))
```

```python
from keras import optimizers

sgd = optimizers.Adam(lr=0.003)
model.compile(sgd,loss='mse')

model.summary()
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint

earlystop = EarlyStopping(monitor ='loss', min_delta =0 ,patience = 50)
```

```python
import time

start=time.time()

history= model.fit(X_train, Y_train, validation_split=0.2, batch_size= 500, epochs= 1000, verbose= 1, callbacks= [earlystop],)

stop= time.time()

print(f"Training time: {stop - start}s")
```

```python
import matplotlib.pyplot as plt

plot_df = pd.DataFrame.from_dict({'train_loss':history.history['loss'], 'Validation_loss' :history.history['val_loss']})
plot_df.plot(logy=True , figsize=(6,6), fontsize=12)
plt.xlabel('Epochs',fontsize=12)
plt.ylabel('Loss',fontsize=12)
plt.savefig('Taining vs Loss curve (CNN-LSTM).pdf')
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true),np.array(y_pred)

    ## Note: does not handle mix 1d representation
    #if _is_1d(y_true): 
    #    y_true, y_pred = _check_1d_array(y_true, y_pred)

    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

MAPE=mean_absolute_percentage_error(rescaled_target_load, rescaled_predicted_load)
print(MAPE)
```
